    #making class to generate raster file
    class Generateraster:
        def __init__(self,model,sourcepath,resultpath,resultname,scaler):
    #     def __init__(self,model,sourcepath,resultpath,resultname):
            self.sourcepath=sourcepath
            self.model=model
            self.scaler=scaler
            self.resultpath=resultpath
            self.resultname=resultname
            raster=rasterio.open(self.sourcepath,nodata=0)
            bands=[5,6,8,10,14]
            rasterdata=raster.read(bands)
            rasterdata=np.array(rasterdata,dtype=np.float64)
            print("readed data.....")
            shape=rasterdata.shape
            # (4, 5490, 10488)
            # -----------------------------------------transposing array inorder to fit in model-------------------------------------------
            rastertranspose=rasterdata.transpose()
            # ---------------------------------------flattening array inorder to pass inside model------------------------------------------------
            rastertranspose=rastertranspose.reshape(-1,len(bands))
            print("Data scaling....")
            rastertranspose=self.scaler.transform(rastertranspose)
            print("Predicting....")
            predictednew=model.predict(rastertranspose)
            # ----------------------converting predicted into list -----------------------------------
            predictedlist=predictednew.tolist()
            newpredicted=[]
            for i in predictedlist:
                if i=='Forest':
                    newpredicted.append(1)
                if i=='Non-Forest':
                    newpredicted.append(2)
            # ----------------------------converting back into array------------------------------------------
            newpredictedarray=np.dstack(newpredicted)
            newpredictedarray1=newpredictedarray.reshape(shape[1],shape[2])#reshaping data
            newpredictedarray2=newpredictedarray1.transpose()
            newpredictedarray2=newpredictedarray1.reshape(shape[2],shape[1])
            newpredictedlist=[newpredictedarray2,newpredictedarray2,newpredictedarray2,newpredictedarray2]
            newnewarray=np.dstack(newpredictedlist)
            newtranspose=newnewarray.transpose()
            # ---------------------------writing raster file -------------------------------------------------------
            driverTiff = gdal.GetDriverByName('GTiff')
            naip_ds = gdal.Open(self.sourcepath)
            clfds = driverTiff.Create(f"{self.resultpath}\\{self.resultname}", naip_ds.RasterXSize, naip_ds.RasterYSize,
                                    1, gdal.GDT_Float32)
            clfds.SetGeoTransform(naip_ds.GetGeoTransform())
            clfds.SetProjection(naip_ds.GetProjection())
            clfds.GetRasterBand(1).WriteArray(newtranspose[0])
            clfds = None
            print(f"Ended writing raster{time.ctime()}")

    #making class to generate raster file
    class Generateraster:
        def __init__(self,model,sourcepath,resultpath,resultname,scaler):
    #     def __init__(self,model,sourcepath,resultpath,resultname):
            self.sourcepath=sourcepath
            self.model=model
            self.scaler=scaler
            self.resultpath=resultpath
            self.resultname=resultname
            raster=rasterio.open(self.sourcepath,nodata=0)
            bands=[1,2,3,4,5]
            rasterdata=raster.read(bands)
            rasterdata=np.array(rasterdata,dtype=np.float64)
            print("readed data.....")
            shape=rasterdata.shape
            # (4, 5490, 10488)
            # -----------------------------------------transposing array inorder to fit in model-------------------------------------------
            rastertranspose=rasterdata.transpose()
            # ---------------------------------------flattening array inorder to pass inside model------------------------------------------------
            rastertranspose=rastertranspose.reshape(-1,len(bands))
            print("Data scaling....")
            rastertranspose=self.scaler.transform(rastertranspose)
            print("Predicting....")
            predictednew=model.predict(rastertranspose)
            # ----------------------converting predicted into list -----------------------------------
            predictedlist=predictednew.tolist()
            newpredicted=[]
            for i in predictedlist:
                if i=='Forest':
                    newpredicted.append(1)
                if i=='Non-Forest':
                    newpredicted.append(2)
            # ----------------------------converting back into array------------------------------------------
            newpredictedarray=np.dstack(newpredicted)
            newpredictedarray1=newpredictedarray.reshape(shape[1],shape[2])#reshaping data
            newpredictedarray2=newpredictedarray1.transpose()
            newpredictedarray2=newpredictedarray1.reshape(shape[2],shape[1])
            newpredictedlist=[newpredictedarray2,newpredictedarray2,newpredictedarray2,newpredictedarray2]
            newnewarray=np.dstack(newpredictedlist)
            newtranspose=newnewarray.transpose()
            # ---------------------------writing raster file -------------------------------------------------------
            driverTiff = gdal.GetDriverByName('GTiff')
            naip_ds = gdal.Open(self.sourcepath)
            clfds = driverTiff.Create(f"{self.resultpath}\\{self.resultname}", naip_ds.RasterXSize, naip_ds.RasterYSize,
                                    1, gdal.GDT_Float32)
            clfds.SetGeoTransform(naip_ds.GetGeoTransform())
            clfds.SetProjection(naip_ds.GetProjection())
            clfds.GetRasterBand(1).WriteArray(newtranspose[0])
            clfds = None
            print(f"Ended writing raster{time.ctime()}")
